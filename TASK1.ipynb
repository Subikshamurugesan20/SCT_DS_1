```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
age_data = np.random.randint(18, 65, 100) 

# Simulated ages between 18 and 65 for 100 people
gender_data = np.random.choice(['Male', 'Female', 'Other'], 100, p=[0.45, 0.45, 0.1])  

# Plotting the histogram for age distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(age_data, bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Plotting the bar chart for gender distribution
plt.subplot(1, 2, 2)
unique, counts = np.unique(gender_data, return_counts=True)
plt.bar(unique, counts, color=['lightblue', 'lightpink', 'lightgreen'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
